```python
import math
import numpy
import seaborn as sns
# sns.color_palette("Set2")
from matplotlib import pyplot as plt
```

```python
timings_search_array = [0.001777 ,0.003403 , 0.009901 , 0.019341 , 0.022548 , 0.028808   ]
timings_search_linkedlist = [0.002872 , 0.00502 , 0.009836 , 0.013091 , 0.014079 , 0.014842 ]
timings_search_trie = [0.000013 , 0.000015 , 0.000015 , 0.000016 , 0.000016 , 0.000018 ]
types = ["array", "linkedlist", "trie"]
num_input = ["25k", "50k", "100k", "150k", "175k", "200k"]
# plt.plot(num_input, timings_search_array, "g-")
# plt.plot(num_input, timings_search_linkedlist, "r-")
plt.plot(num_input, timings_search_trie, "b-")
plt.xlabel('Number of elements')
plt.ylabel('Time per operation (ms)')
plt.title("Search Word Function")
plt.title('Search word function')
plt.legend(types, loc = "upper left")
# plt.figure(dpi=1200)
plt.show()
```

```python
sns.lineplot(x=num_input, y=timings_search_array, label="array")
sns.lineplot(x=num_input, y=timings_search_linkedlist, label="linkedlist")
sns.lineplot(x=num_input, y=timings_search_trie, label="trie").set(xlabel ="Input Size", ylabel ="Time Taken(in nanoseconds)", title ="Search Word Function")
sns.set(rc={"figure.dpi":300})
```
